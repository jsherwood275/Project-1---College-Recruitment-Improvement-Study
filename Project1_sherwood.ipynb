```python
#Add a new column names year by taking the first 4 digits of the timestamp.  
df['year'] = df['timestamp'].str[:4]

#Change timestamp column to a timestamp data type.
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y/%m/%d %I:%M:%S %p %Z')
```

```python
#Clean the race ethnicity values
# Replace those values with 'Multi-Racial'
race_ethnicity_rows = df['race_ethnicity'].str.contains(';', case=False)
df.loc[race_ethnicity_rows, 'race_ethnicity'] = 'Multi-Racial'
```

```python
#search for engineering majors and include them in the overall engineering major.
engineering_rows = df['degree'].str.contains('engineering', case=False)
df.loc[engineering_rows, 'degree'] = 'Engineering'
```

```python
#Clean the degree values
# Get the values where count is less than or equal to 5
counts = df['degree'].value_counts()
filtered_counts = counts[counts <= 5]
values_to_replace = filtered_counts.index
# Replace those values with 'Other'
df.loc[df['degree'].isin(values_to_replace), 'degree'] = 'Other'
```

```python
#Dropping these columns as they do not align to recruitment.
df.drop(axis=1, columns=['additional_comp_classes','reason','motivation_all'],inplace=True)
```

```python
#replace all NaN with N/A
df.fillna('N/A', inplace=True)
```
