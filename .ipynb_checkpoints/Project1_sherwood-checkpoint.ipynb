```python
#check the columns in each data set to reduce to common set of columns in all CSV

#Fix Age Columns to ensure they are common across all DF's.  
#There is an extra space in a few of the CSV
for df in [df_2020, df_2021, df_2022, df_2023]:
    df.rename(columns={'Age ':'Age'}, inplace=True)

def indexColumns(DF):
    return set(DF.columns)  # Return a set of column names

# Call the function with your DataFrame and store the result in a variable
yr2020Columns = indexColumns(df_2020)
yr2021Columns = indexColumns(df_2021)
yr2022Columns = indexColumns(df_2022)
yr2023Columns = indexColumns(df_2023)

#create a set to pull common columns
common_columns = yr2020Columns.intersection(yr2021Columns, yr2022Columns, yr2023Columns)

# Find the unique columns not shared among all DFs
unique_columns = yr2020Columns ^ yr2021Columns ^ yr2022Columns ^ yr2023Columns

# Convert the set to a list if needed
common_columns_list = list(common_columns)
unique_columns_list = list(unique_columns)

# Filter each DataFrame to keep only the common columns
df_2020 = df_2020[common_columns_list]
df_2021 = df_2021[common_columns_list]
df_2022 = df_2022[common_columns_list]
df_2023 = df_2023[common_columns_list]

#check to see of column counts match
print(df_2020.shape)
print(df_2021.shape)
print(df_2022.shape)
print(df_2023.shape)
```

```python
#create a list to store all datasets to Concatenate all DataFrames into a single DataFrame
dfs = [df_2020, df_2021, df_2022,df_2023]
df = pd.concat(dfs, ignore_index=True)
```

```python
#Add a new column names year by taking the first 4 digits of the timestamp.  
df['year'] = df['timestamp'].str[:4]

#Change timestamp column to a timestamp data type.
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y/%m/%d %I:%M:%S %p %Z')
```

```python
#Clean the race ethnicity values
# Replace those values with 'Multi-Racial'
race_ethnicity_rows = df['race_ethnicity'].str.contains(';', case=False)
df.loc[race_ethnicity_rows, 'race_ethnicity'] = 'Multi-Racial'
```

```python
#search for engineering majors and include them in the overall engineering major.
engineering_rows = df['degree'].str.contains('engineering', case=False)
df.loc[engineering_rows, 'degree'] = 'Engineering'
```

```python
#Clean the degree values
# Get the values where count is less than or equal to 5
counts = df['degree'].value_counts()
filtered_counts = counts[counts <= 5]
values_to_replace = filtered_counts.index
# Replace those values with 'Other'
df.loc[df['degree'].isin(values_to_replace), 'degree'] = 'Other'
```

```python
# Replace 'a few strings in Comp major first choice' with 'First Choice'
df['comp_first_choice'].replace('did not swap', 'First Choice', inplace=True)
df['comp_first_choice'].replace('I did not change my major', 'First Choice', inplace=True)
```

```python
# Replace '18 and younger' with '18-'
df['age'].replace('18 and younger', '18 -', inplace=True)
```

```python
# Drop records where 'course_enrolled' is blank
df.dropna(subset=['course_enrolled'], inplace=True)
```

```python
#Question 2
#2.	What is the demographic breakdown of applicants to the CCM IT program over the past 3 years, 
#   and are there any noticeable trends or shifts in gender, ethnicity, or age?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Group the data by year and demographic variables
grouped_data = df.groupby(['year', 'gender', 'race_ethnicity', 'age'])

#Calculate counts for each demographic variable within each year
demographic_counts = grouped_data.size().reset_index(name='count')

#Visualize the demographic breakdown over the past 4 years
plt.figure(figsize=(12, 8))
sns.countplot(data=demographic_counts, x='year', hue='gender')
plt.title('Gender Distribution of survey data over the Past 4 Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Gender', loc=0)
plt.show()

plt.figure(figsize=(12, 8))
sns.countplot(data=demographic_counts, x='year', hue='race_ethnicity')
plt.title('Ethnicity Distribution of survey data Over the Past 4 Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Ethnicity', loc=0)
plt.show()

plt.figure(figsize=(12, 8))
sns.histplot(data=demographic_counts, x='age', hue='year', multiple='stack', bins=20)
plt.title('Age Distribution of survey data Over the Past 4 Years')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(labels=["2020","2021","2022","2024"],title='Year')
plt.show()
```

```python
#Question 4
#4.	Look at the records of just the Computing Majors only as one group. 
#Computing Majors include: Computer Science, IT, Game Dev, Digital Media Technology, Technical Studies, Information Security, 
#Data Analytics and Web Development. 
#Show percentage of these who computing major their first choice and the percentage who it was not their first choice? 
#Use a pie chart to show the Computing Majors vs Other.

import pandas as pd
import matplotlib.pyplot as plt

# List of computing majors
computing_majors = [
    'CIS Game Development Option',
    'CIS Game Development Option, Applied Associate of Science',
    'Computer Science',
    'Computer Science, Associate of Science',
    'Data Analytics Certificate of Achievement',
    'Digital Media Technology',
    'Information Security Certificate of Achievement',
    'Information Technology',
    'Information Technology, Applied Associate of Science',
    'Technical Studies, Applied Associate of Science',
    'Web Development Certificate of Achievement'
]

# Filter the data for computing majors
computing_major_data = df[df['degree'].isin(computing_majors)]

# Count the total number of computing majors
total_computing_majors = len(computing_major_data)

# Count the number of computing majors where it was their first choice
first_choice_count = (computing_major_data['comp_first_choice'] == 'First Choice').sum()

# Calculate the percentage of computing majors where it was their first choice
first_choice_percentage = (first_choice_count / total_computing_majors) * 100

# Calculate the percentage of computing majors where it was not their first choice
not_first_choice_percentage = 100 - first_choice_percentage

# Create a pie chart
labels = ['Computing Majors - First Choice', 'Computing Majors - Not First Choice']
sizes = [first_choice_percentage, not_first_choice_percentage]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # explode the first slice (to highlight it)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Computing Majors as First Choice vs Not First Choice')
plt.show()


```
